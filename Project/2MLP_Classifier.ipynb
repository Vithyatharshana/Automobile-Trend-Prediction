```python
import io
df = pd.read_csv(io.BytesIO(uploaded['Processed_Data.csv']))
df.drop(df.iloc[:, 0:1], inplace=True, axis=1)
df
```

```python
from sklearn import datasets
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:, 0:5].values 
y = df["Sale_2"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
projectMLP = MLPClassifier(solver='lbfgs', activation='logistic', hidden_layer_sizes=(4,), random_state=1)
projectMLP.fit(x_train, y_train)
print("Accuracy: ",projectMLP.score(x_test, y_test))
projectMLP.predict(x_test)
```

```python
print(projectMLP)
```

```python
expected_y  = y_test
predicted_y = projectMLP.predict(x_test)
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))
```
