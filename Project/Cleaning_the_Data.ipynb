```python
import numpy as np
import pandas as pd
```

```python
# Taking the dataset 
import io
old_df = pd.read_csv(io.BytesIO(uploaded['Data.csv']))
old_df
```

```python
# Replacing all the , to . for computations
old_df['Male'] = old_df['Male'].str.replace(',','')
old_df['Female'] = old_df['Female'].str.replace(',','')
old_df['Unknown'] = old_df['Unknown'].str.replace(',','')
old_df['Total'] = old_df['Total'].str.replace(',','')
old_df
```

```python
# Converting string to int for computations
# One - hot encoding is done as machine understands only numbers
old_df["Fuel"]=old_df["Fuel"].map({"gasoline":1,"diesel":2})
old_df["Fuel"]
```

```python
# So here Gasoline is 0 and Fuel is 1
old_df = pd.get_dummies(data=old_df,columns=["Fuel"],drop_first=True)
old_df
```

```python
old_df.to_csv('Separated_Data.csv')
```

```python
df = pd.read_csv('Separated_Data.csv')
df
```

```python
def sales(define):
    if(int(define['Total']>62000)):
        return 'Trending'
    else:
        return 'Not trending'
df['Sale'] = df.apply(sales, axis=1)
df
```

```python
df["Sale"] = df["Sale"].map({"Trending":1,"Not trending":2})
df["Sale"]
```

```python
df = pd.get_dummies(data = df,columns=["Sale"],drop_first=True)
df
```

```python
df.to_csv('Separated_Data2.csv')
# Shifted Fuel_2.0 near Engine CC manually for easy calculation and named the file as "Processed_Data"
```
