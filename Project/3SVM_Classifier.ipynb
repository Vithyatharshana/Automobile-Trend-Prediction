```python
# Applying Min - Max Normalization to the dataset
import io
df = pd.read_csv(io.BytesIO(uploaded['Processed_Data.csv']))
df.drop(df.iloc[:, 0:1], inplace=True, axis=1)
df
```

SVMs assume that the data it works with is in a standard range, usually either 0 to 1, or -1 to 1 (roughly). **So the normalization of feature vectors prior to feeding them to the SVM is very important**

```python
# Applying Min - Max Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)
print(normalized_data)
```

```python
# Putting the values obtained in a tabular format
nor_df = pd.DataFrame(normalized_data, columns = df.columns)
print(nor_df)
```

```python
nor_df.to_csv('Normalized_Data.csv')
df = pd.read_csv(r'Normalized_Data.csv')
df.drop(df.iloc[:, 0:1], inplace=True, axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,0:5]
y = df["Sale_2"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear') # for actual dataset, radial func has more accuracy
clf.fit(x_train, y_train)
```

```python
clf.support_vectors_
```

```python
Acc = clf.score(x,y)
print("Accuracy: ", Acc)
```

```python
clf.predict(x_test)
```

```python
# Testing Accuracy of other kernel functions
# Linear function
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(x_train, y_train)
clf.score(x,y)
```

```python
# Polynomial function
from sklearn import svm
clf = svm.SVC(kernel='poly')
clf.fit(x_train, y_train)
clf.score(x,y)
```

```python
# Radial Basis function
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(x_train, y_train)
clf.score(x,y)
```

```python
# Sigmoid function
from sklearn import svm
clf = svm.SVC(kernel='sigmoid')
clf.fit(x_train, y_train)
clf.score(x,y)
```
