```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train.values.reshape(-1,1), Y_train)
```

```python
Y_pred = regr.predict(X_test.values.reshape(-1,1))
Y_pred
```

```python
print("Mean squared error: %.2f" % mean_squared_error(Y_test, Y_pred))
```
