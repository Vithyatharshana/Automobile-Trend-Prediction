Skewness & Kurtosis

Skewness: Measure of the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images. A distribution can have right (or positive), left (or negative), or zero skewness.

Kurtosis: Measure of the tailedness of a distribution. Tailedness is how often outliers occur. Excess kurtosis is the tailedness of a distribution relative to a normal distribution. Distributions with medium kurtosis (medium tails) are mesokurtic. Distributions with low kurtosis (thin tails) are platykurtic

We applied Skewness and Kurtosis for all columns in the Dataset

Data Sampling
Data sampling is a statistical analysis technique used to select, process, and analyze a representative subset of a population. It is also used to identify patterns and extrapolate trends in an overall population.

Simple Random Sampling and Systematic Sampling were applied on our Dataset.

•	Simple Random Sampling: One of the best probability sampling techniques that helps in saving time and resources. Single member of a population is chosen randomly, merely by chance. Each individual has the same probability of being chosen to be a part of a sample.

•	Systematic Sampling: Here the selection of elements is systematic and not random except the first element. 
Elements of a sample are chosen at regular intervals of population. All the elements are put together in a sequence first where each element has the equal chance of being selected

```python
# Data Sampling
# Simple Random Sampling
simple_random_sample = df.sample(n=4).sort_values(by='Fuel_2.0')
simple_random_mean = round(simple_random_sample['Price'].mean(),3)
print(simple_random_mean)
simple_random_sample
```

```python
# Systematic Sampling
def systematic_sampling(df, step):
    
    indexes = np.arange(0,len(df),step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

systematic_sample = systematic_sampling(df, 5)
systematic_mean = round(systematic_sample['Price'].mean(),3)
systematic_sample
 
```
