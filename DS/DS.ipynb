```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Data.csv')
df
```

```python
q1 = df['Power'].quantile(0.25)
q3 = df['Power'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
```

```python
df['Power'][((df['Power']<lower)|(df['Power']>upper))]
#outliers
```

```python
q1 = df['Engine CC'].quantile(0.25)
q3 = df['Engine CC'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
df['Engine CC'][((df['Engine CC']<lower)|(df['Engine CC']>upper))]
#outliers
```

```python
a=df['Engine CC'].mean()
b=df['Engine CC'].std()
df.loc[df['Engine CC']<lower,'Engine CC']=a-3*b
df.loc[df['Engine CC']>upper,'Engine CC']=a+3*b
#replacing outliers
```

```python
q1 = df['Transmission'].quantile(0.25)
q3 = df['Transmission'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
df['Transmission'][((df['Transmission']<lower)|(df['Transmission']>upper))]
#outliers
```

```python
a=df['Transmission'].mean()
b=df['Transmission'].std()
df.loc[df['Transmission']<lower,'Transmission']=a-3*b
df.loc[df['Transmission']>upper,'Transmission']=a+3*b
#replacing outliers
```

```python
q1 = df['Price'].quantile(0.25)
q3 = df['Price'].quantile(0.75)
iqr = q3-q1
lower = q1-1.5*iqr
upper = q3+1.5*iqr
df['Price'][((df['Price']<lower)|(df['Price']>upper))]
#outliers
```

```python
a=df['Price'].mean()
b=df['Price'].std()
df.loc[df['Price']<lower,'Price']=a-3*b
df.loc[df['Price']>upper,'Price']=a+3*b
#replacing outliers
```

```python
df.isna().sum()
#null values
```

```python
df['Male'] = df['Male'].str.replace(',','')
df['Female'] = df['Female'].str.replace(',','')
df['Unknown'] = df['Unknown'].str.replace(',','')
df['Total'] = df['Total'].str.replace(',','')
df
#replacing , with '' to make the attributes integers
```

```python
df.to_csv('Cleaned_data.csv')
```

```python
# Applying Min - Max Normalization to the dataset
import io
df = pd.read_csv(io.BytesIO(uploaded['Processed_Data.csv']))
df.drop(df.iloc[:, 0:1], inplace=True, axis=1)
df
```

```python
# Applying Min - Max Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df.copy())
print(normalized_data)
```

```python
# Z-score normalization
df_z_scaled = df.copy()
for column in df_z_scaled.columns:
    df_z_scaled[column] = (df_z_scaled[column] -
                           df_z_scaled[column].mean()) / df_z_scaled[column].std()     
display(df_z_scaled)
```
